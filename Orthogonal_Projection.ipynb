# Maximum Likelihood as Orthogonal Projection

Consider the basic univariate linear regression problem: $$y=x\theta+\epsilon, \quad\epsilon\sim\mathcal{N}(0,\sigma^2).\newline$$ If we have a training set $\{(x_1,y_1),...,(x_N,y_N)\}$ and we let $X\in\mathbb{R}^N$ and $\mathbf{y}\in\mathbb{R}^N$, then we can express the maximum likelihood estimator as follows: $$\newline$$ $$\theta_{ML}=(X^TX)^{-1}X^T\mathbf{y},\newline$$ $$X\theta_{ML}=\frac{XX^T}{X^TX}\mathbf{y}.$$

```python
import numpy as np
import matplotlib.pyplot as plt

sample_size = 20
x_pts = np.sort(np.random.uniform(-4, 4, sample_size))
y_pts = (1/3) * x_pts + np.random.normal(0, 0.2, sample_size)

fig, ax = plt.subplots()
ax.set_title("Random Sample")
ax.set_ylim([-4, 4])
sampled, = ax.plot(x_pts, y_pts, 'b.', label="sampled")
```

```python
θ_mle = x_pts@y_pts/(x_pts@x_pts)
x_mle = np.linspace(-4,4,200)
y_mle = θ_mle * x_mle

y_obs_mle = θ_mle * x_pts

fig, ax = plt.subplots()
ax.set_title("Random Sample")
ax.set_ylim([-4, 4])
sampled, = ax.plot(x_pts, y_pts, 'b.', label="Observations")
mle, = ax.plot(x_mle, y_mle, color='black', label="Maximum Likelihood Estimate")
# TODO: Draw orange lines connecting observed data and projections
print(np.vstack([y_pts, y_obs_mle]).shape)
print(np.vstack([x_pts, x_pts]).shape)
proj, = ax.plot(np.vstack([x_pts, x_pts]), np.vstack([y_pts, y_obs_mle]), '.', color='orange', label="Projections")
ax.legend(handles=[sampled, mle, proj], loc="best")
```
